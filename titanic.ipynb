```python
# 891 passengers is a moderate dataset.
# 15 features is good amount of information for ML tasks.
# Age has 177 missing values ,it requires imputation
# Columns include numerical, categorical, boolean it requires preprocessing
# Some columns (class, deck) are categorical types already
# The survived column shows:0 → did not survive ,1 → survived
```

```python
missing_columns = titanic.columns[titanic.isnull().any()]
print(missing_columns)
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
titanic = titanic.drop(columns=['Cabin'])
```

```python
# Machine Learning models require numerical input,Prevents mathematical errors,
# One-hot encoding avoids false ordering
```

```python
titanic.head()
```

```python
# Feature scaling is an essential preprocessing step because:
# It ensures all features contribute equally
# Scaling improves performance for: KNN,K-Means,SVM,Neural

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

scaler = StandardScaler()

titanic[['age_scaled_std', 'fare_scaled_std']] = scaler.fit_transform(titanic[['Age', 'Fare']])
minmax = MinMaxScaler()
titanic[['age_scaled_mm', 'fare_scaled_mm']] = minmax.fit_transform(titanic[['Age', 'Fare']])
```

```python
comparison = titanic[['Age', 'Fare', 'age_scaled_std', 'fare_scaled_std', 'age_scaled_mm', 'fare_scaled_mm']].head()
print(comparison)
```

```python
titanic['family_size'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic['is_child'] = titanic['Age'] < 18
```

```python
# family_size gives the model additional structure:
# Alone → likely to be ignored
# Small family → helped by social support
# Large family → overcrowded, harder to evacuate
# is_child (age < 18) shows women and children first was the evacuation priority on the Titanic.
```

```python
columns_to_drop = ['Ticket', 'Fare', 'Cabin', 'Embarked']

# Drop only if column exists to avoid errors
titanic = titanic.drop(columns=[col for col in columns_to_drop if col in titanic.columns])
```

```python
titanic = titanic.drop(columns=['SibSp', 'Parch',])
```

```python
titanic.drop(columns=['Age'], inplace=True)
titanic.head()
```
